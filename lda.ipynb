```python
# Create dictionaries of the words from the subtitles
id2word = corpora.Dictionary(texts)

# Construct the TDFs (Term Document Frequencies) for each text
corpus = [id2word.doc2bow(text) for text in texts]
```

```python
# Creating/loading model from dataset
number_of_topics = 10
lda = LdaMulticore(corpus=corpus, id2word=id2word, num_topics=number_of_topics)
```

```python
# Print all topics with top 10 keywords
pprint(lda.print_topics(number_of_topics))
```

```python
topic_visualization = gensimvis.prepare(lda, corpus, id2word)
```
